```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import pandas as pd
import glob
```

```python
def get_histories():
    pickle_files = sorted(glob.glob("*.pkl"))
    datas = {}
    for fname in pickle_files:
        name = fname.split('.')[0]
        with open(fname, 'rb') as f:
            datas[name] = pickle.load(f)
        
    return datas
```

```python
datas = get_histories()
```
